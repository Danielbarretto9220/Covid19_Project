Importing Libraries

```python
import warnings
import itertools
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
plt.style.use('fivethirtyeight')
import pandas as pd
import statsmodels.api as sm
import matplotlib
```

```python
matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
```

```python
#TO read csv file
da=pd.read_csv("Dakshina Kannada.csv")
da.head()
```

```python
da.shape
```

```python
da['Date'] = pd.to_datetime(da['Date'])
da['Date'] = da['Date'].dt.strftime('%d/%m/%Y')
da['Date'] 
```

```python
df = pd.DataFrame(da, columns = ['Date','Today’s Positives'])
df1 = pd.DataFrame(da, columns = ['Date','Today’s Discharges'])
# Set the Date as Index
df.index = da['Date']
del df['Date']
df1.index = df1['Date']
del df1['Date']


df.plot(figsize=(12,4), color='r')
df1.plot(figsize=(12,4), color='g')
plt.show()
```

```python
df = pd.DataFrame(da, columns = ['Date','Total Positives'])
df1 = pd.DataFrame(da, columns = ['Date','Total Discharges'])
# Set the Date as Index
df.index = da['Date']
del df['Date']
# Set the Date as Index
df1.index = df1['Date']
del df1['Date']

df.plot(figsize=(12,4), color='r')
df1.plot(figsize=(12,4), color='g')
plt.show()
```

```python
df = pd.DataFrame(da, columns = ['Date','Total Active Cases'])

# Set the Date as Index
df.index = da['Date']
del df['Date']


df.plot(figsize=(12,6))
plt.show()
```

```python
df = pd.DataFrame(da, columns = ['Date','Total Covid Deaths'])

# Set the Date as Index
df.index = df['Date']
del df['Date']


df.plot(figsize=(15, 6))
plt.show()
```

```python
df = pd.DataFrame(da, columns = ['Date','Death due to Non- Covid reasons'])

# Set the Date as Index
df.index = df['Date']
del df['Date']


df.plot(figsize=(15, 6))
plt.show()
```
