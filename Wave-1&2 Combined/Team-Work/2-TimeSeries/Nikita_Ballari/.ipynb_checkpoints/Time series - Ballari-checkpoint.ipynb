```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Ballari Wave2.xlsx')
data
```

```python
data.columns
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data['Date'] = data['Date'].dt.strftime('%d/%m/%Y')
data['Date'] 
```

```python
df = pd.DataFrame(data, columns = ['Date','Today’s Positives'])
df1 = pd.DataFrame(data, columns = ['Date','Today’s Discharges'])
# Set the Date as Index
df.index = df['Date']
del df['Date']
df1.index = df1['Date']
del df1['Date']


df.plot(figsize=(12,4), color='r')
df1.plot(figsize=(12,4), color='g')
plt.show()
```

The curve for positive cases and discharges per day were both peaked between 10 may to 20 may but the discharge rate was comparitively high(3500) than rate of positive cases(2500). After May 20 both the curves follow same number of frequency.

```python
df = pd.DataFrame(data, columns = ['Date','Total Positives'])
df1 = pd.DataFrame(data, columns = ['Date','Total Discharges'])
# Set the Date as Index
df.index = df['Date']
del df['Date']
# Set the Date as Index
df1.index = df1['Date']
del df1['Date']

df.plot(figsize=(12,4), color='r')
df1.plot(figsize=(12,4), color='g')
plt.show()
```

The curve was flat lined before 2nd week of april and started going up after that so we can consider the second wave to start from 3rd week of april in Ballari District.
As of now, the total positive cases are 90000 and total discharges are 80000. Both the curves are elevating almost at same frequency at the same time.

```python
df = pd.DataFrame(data, columns = ['Date','Total Active Cases'])

# Set the Date as Index
df.index = df['Date']
del df['Date']


df.plot(figsize=(12,6))
plt.show()
```

The highest number of cases were around 20 May  of frequency of 20000 and decreasing in a faster rate since 15 days and have reached the frequency of 7500 cases

```python
df = pd.DataFrame(data, columns = ['Date','Today’s Reported Covid Deaths'])

# Set the Date as Index
df.index = df['Date']
del df['Date']


df.plot(figsize=(15, 6))
plt.show()
```

There is no sign of the curve to be gradually increasing or decreasing between 10th April to 25th of May, the pattern is not following a single direction and is frequently fluctuating in every few days. From the past 10-12 days the deaths have reached to the frequency of 5-10 people each day.

```python
df = pd.DataFrame(data, columns = ['Date','Total Covid Deaths'])

# Set the Date as Index
df.index = df['Date']
del df['Date']


df.plot(figsize=(15, 6))
plt.show()
```

The deaths have suddenly started increasing around last week of April and its growing at the same rate after that. Till now 1400 deaths have occured in Ballari District.

```python
df = pd.DataFrame(data, columns = ['Date','Death due to Non- Covid reasons'])

# Set the Date as Index
df.index = df['Date']
del df['Date']


df.plot(figsize=(15, 6))
plt.show()
```
