```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mysuru.csv")
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)

df
```

```python
df = df.rename(columns = {'District Name':'District','Today’s Positives':'Todays_Positives','Total Positives':'Total_Positives','Today’s Discharges':'Todays_Discharges','Total Discharges':'Total_Discharges','Total Active Cases':'Total_Active_Cases','Today’s Reported Covid Deaths':'Todays_Reported_Covid_Deaths','Total Covid Deaths':'Total_Covid_Deaths','Death due to Non- Covid reasons':'Death_due_to_Non-Covid_reasons'})
```

```python
df['Date']=pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
df['Date']
```

```python
df['Todays_Positives'].unique()
```

```python
df["Todays_Positives"].replace({"1331#": "1331"}, inplace=True)
```

```python
df['Todays_Positives']=df['Todays_Positives'].astype(str).astype(float)
```

```python
df.dtypes
```

```python
df0 = pd.DataFrame(df, columns = ['Date','Total_Active_Cases'])
# Set the Date as Index
df0.index = df0['Date']
del df0['Date']
df0.plot(figsize=(12,6))
plt.show()
```

The highest number of cases were found around 2nd week of october at the frequency of 8300 and decreasing in a faster rate seen from last week of october 15 days and have reached the frequency of 500 cases in the 2nd week of november and remains the same till the 2 nd week of april.The highest number of cases were found around last week May of frequency of 17500.

```python
df1 = pd.DataFrame(df, columns = ['Date','Todays_Positives'])
# Set the Date as Index
df1.index = df1['Date']
del df1['Date']
df1.plot(figsize=(12,6),color='r')
plt.show()

df2 = pd.DataFrame(df, columns = ['Date','Todays_Discharges'])
# Set the Date as Index
df2.index = df2['Date']
del df2['Date']
df2.plot(figsize=(12,6),color='g')
plt.show()

df3 = pd.DataFrame(df, columns = ['Date','Todays_Reported_Covid_Deaths'])
# Set the Date as Index
df3.index = df3['Date']
del df3['Date']
df3.plot(figsize=(12,6),color='y')
plt.show()
```

1.The curve for positive cases per day is at it's peak 10 May to 20 th May it is started decreasing from December and has start increasing from the month of mid of April 2021 and reached to it's highest peak in Mid of the May month.

2.Similarly Curve for the Discharge cases Per Day is higher in the mid month of (10-20)May,it is showing the similar trend in  other months also.

3.The curve for the reported death per day is following the same trend as above,the highest death cases are in the last days of August and mid days of the May month.

```python
df1 = pd.DataFrame(df, columns = ['Date','Total_Positives'])
# Set the Date as Index
df1.index = df1['Date']
del df1['Date']
df1.plot(figsize=(12,6),color='r')
plt.show()

df2 = pd.DataFrame(df, columns = ['Date','Total_Discharges'])
# Set the Date as Index
df2.index = df2['Date']
del df2['Date']
df2.plot(figsize=(12,6),color='g')
plt.show()

df3 = pd.DataFrame(df, columns = ['Date','Total_Covid_Deaths'])
# Set the Date as Index
df3.index = df3['Date']
del df3['Date']
df3.plot(figsize=(12,6),color='y')
plt.show()
```

The curve started going up from frist week of july which is at its highest peak in the first week of november(5000) and after that it is flat till 3rd week of april, so we can consider the second wave to start from 3rd week of april in Mysuru District.
As of now, the total positive cases are 1,40,000 and total discharges are 1,20,000 and the total covid deaths are 1500 till now.all the curves are showing almost same frequency at the same time.
