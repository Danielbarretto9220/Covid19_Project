```python
columns = list(bagalakote.columns)
ind = columns.index('date')
columns[0]
```

```python
def plotting(dis_df, dis_name, feature):
    columns = list(dis_df.columns)
    ind = columns.index(f'{feature}')
    plt.figure(figsize = (30, 20))
    figure = dis_df[feature].plot(kind = 'bar', colormap = 'Reds')
    figure2 = dis_df[feature].plot(kind = 'line')
    figure.set_xticklabels(labels = dis_df.date, rotation=90)
    plt.title(f'{columns[ind]}: {dis_name}')
    plt.show()
```

```python
plotting(bagalakote, 'bagalakote', 'Todays_Positives')
```
